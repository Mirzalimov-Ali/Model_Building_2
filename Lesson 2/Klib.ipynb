```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[7, 7])
color_counts = df['Color'].value_counts()
plt.pie(
    color_counts,
    labels=color_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.show()
```
